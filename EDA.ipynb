```python
# importar funciones
from utils.def_classify_location import get_group_location
from utils.def_region_classification import get_region_classification
```

```python
# Llamar a las funciones creadas en utils, para la selección de la región de estudio.
df_avocado_hotels = df_avocado.copy()

# Crear la columna 'region_type'
df_avocado_hotels['region_type'] = df_avocado_hotels['region'].apply(get_region_classification)
# Crear la columna 'location'
df_avocado_hotels['location'] = df_avocado_hotels['region'].apply(get_group_location)
```

```python
df_avocado_hotels['location'].unique()
```

```python
# Filtrar el dataset por la region de estudio
df_avocado_hotels = df_avocado_hotels[df_avocado_hotels["location"].isin(['newyork', 'texas', 'louisiana', 'florida', 'california',])]
```

```python
df_avocado_hotels.head(10)
```

```python
# Agregamos nuevas columnas al DataFrame df_avocado_hotels

# Calculo la cantidad de aguacates en función del tipo
df_avocado_hotels['cantidad_aguacates'] = df_avocado_hotels.apply(
    lambda row: row['Total Volume'] / (0.25 if row['type'] == 'conventional' else 0.3),
    axis=1
)

# Calculo cantidad de toneladas
df_avocado_hotels['cant_toneladas'] = (df_avocado_hotels['Total Volume'] / 1000).round(2)

# Calcular el precio por tonelada
df_avocado_hotels['precio_tonelada'] = (
    (df_avocado_hotels['AveragePrice'] * df_avocado_hotels['cantidad_aguacates']) / df_avocado_hotels['cant_toneladas']
)
```

```python
# Convertir columna date a datetime
df_avocado_hotels['Date'] = pd.to_datetime(df_avocado_hotels['Date'])

# Elección de las columnas de interés para el análisis
df_avocado_hotels = df_avocado_hotels[['Date', 'AveragePrice', 'Total Volume', '4046', '4225',
       '4770','type', 'year', 'region','location','cant_toneladas', 'precio_tonelada']]
df_avocado_hotels.columns
```
