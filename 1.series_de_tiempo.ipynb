### 1. **Análisis de Series de Tempo**
Realizar un análisis exhaustivo de las tendencias y patrones temporales presentes en los datos.


```python
import numpy as np
import pandas as pd

from utils.def_classify_location import get_group_location
from utils.def_region_classification import get_region_classification

import matplotlib.pyplot as plt
%config InlineBackend.figure_format = "retina"

import seaborn as sns

#import statsmodels.api as sm
#import statsmodels.formula.api as smf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Carga el dataset
df_orig = pd.read_csv("./data/avocado.csv")
df_orig
```

```python
# Copia del dataframe para preservar datos originales
df = df_orig.copy()
```

#### **Descomponemos Series Temporales de Precios:**

   - **Uso de Datos:** Usa la columna `AveragePrice` y `Date`.
   - **Esperado:** Utiliza la función `seasonal_decompose` de la librería `statsmodels` para descomponer la serie temporal de precios en componentes de tendencia, estacionalidad y ruido. 
     - Convertimos `Date` a tipo datetime usando `pd.to_datetime()`.
     - Agrupamos los datos por `Date` y calcula el promedio de `AveragePrice` utilizando `groupby()` si es necesario.
     - Visualizamos los componentes descompuestos usando `matplotlib` para cada uno de ellos.


```python

# Crear la columna 'region_type'
df['region_type'] = df['region'].apply(get_region_classification)
# Crear la columna 'location'
df['location'] = df['region'].apply(get_group_location)

df['location'].unique()

```

```python
# Convierte 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Filtrar el dataset por la region de estudio
df_avocado_hotels = df[df["location"].isin(['newyork', 'texas', 'louisiana', 'florida', 'california',])]

df_avocado_hotels.head(10)
```

```python
# Agregamos nuevas columnas al DataFrame df_avocado_hotels

# Calculo la cantidad de aguacates en función del tipo
df_avocado_hotels['cantidad_aguacates'] = df_avocado_hotels.apply(
    lambda row: row['Total Volume'] / (0.25 if row['type'] == 'conventional' else 0.3),
    axis=1
)

# Calculo cantidad de toneladas
df_avocado_hotels['cant_toneladas'] = (df_avocado_hotels['Total Volume'] / 1000).round(2)

# Calcular el precio por tonelada
df_avocado_hotels['precio_tonelada'] = (
    (df_avocado_hotels['AveragePrice'] * df_avocado_hotels['cantidad_aguacates']) / df_avocado_hotels['cant_toneladas']
)
```

```python
#Para poder escoger un número de period que tenga sentido en seasonal_decompose primero vamos a entender qué datos tengo

# Crear las columnas 'año', 'mes' y 'día'
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Crear una columna 'dia_semana' que tenga el nombre del día de la semana
df['dia_semana'] = df['Date'].dt.day_name()

# Visualizar el resultado
print(df[['Date', 'year', 'Month','dia_semana']].head())

#Ver qué días tengo registros de entrada en mi tabla
df['dia_semana'].unique()
```

Solo tenemos datos registrados en domingo, por lo tanto los números que tendrán sentido poner en period son:
- 52 para ver una estacionalidad anual
- 26 para ver una estacionalidad semestral (cada 6 meses)
- 13 para ver una estacionalidad trimestral (cada 3 meses)

```python
# Group data by Date to study temporal series
gd_4_temp_series = df[['Date', 'AveragePrice']].groupby('Date').mean()
gd_4_temp_series
```

```python
# Genera la serie temporal
# [period: Expected samples in a full cycle / repetition of the seasonality component]
decomposition = seasonal_decompose(gd_4_temp_series['AveragePrice'], model='additive', period=52)
```

```python
# Dibuja los componentes de la descomposición
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
fig.suptitle('Descomposición en series temporales de \'AveragePrice\'')
decomposition.observed.plot(ax=ax1, title='Observed', ylabel='AveragePrice')
decomposition.trend.plot(ax=ax2, title='Trend', ylabel='AveragePrice')
decomposition.seasonal.plot(ax=ax3, title='Seasonal', ylabel='AveragePrice')
decomposition.resid.plot(ax=ax4, title='Residual', ylabel='AveragePrice')

plt.tight_layout()
plt.show()
```

Se puede observar que anualmente se repite una bajada en el AveragePrice en el mes de Febrero así como un pico de subida en los meses de Octubre y Noviembre.
En septiembre se empieza a acabar la temporada de cosecha del aguacate, y por eso tenemos un costo más alto en Octubre y Noviembre, si le sumamos también que en Noviembre hay el día de Acción de Gracias donde es muy común el uso del aguacate para hacer Guacamole en USA.
